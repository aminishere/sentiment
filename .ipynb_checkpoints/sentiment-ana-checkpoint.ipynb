```python
import numpy as np
import pandas as pd
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
nltk.download('stopwords')
```

```python
data = pd.read_csv('Restaurant_Reviews.tsv', delimiter='\t', quoting=3)
```

```python
print("Dataset shape:", data.shape)
print("Columns:", data.columns)
print(data.head())
```

```python
corpus = []
ps = PorterStemmer() 
```

```python
for i in range(len(data)):
    # Remove all non-alphabetic characters
    review = re.sub('[^a-zA-Z]', ' ', data['Review'][i])

    # Convert to lowercase
    review = review.lower()

    # Split into individual words
    review_words = review.split()

    # Remove stopwords (common words like 'the', 'is', etc.)
    review_words = [word for word in review_words if word not in set(stopwords.words('english'))]

    # Stem words (e.g., "loved" -> "love", "running" -> "run")
    review_words = [ps.stem(word) for word in review_words]

    # Join words back into a single string
    review = ' '.join(review_words)

    # Add cleaned review to corpus
    corpus.append(review)
```

```python
cv = CountVectorizer(max_features=1500)  # Limit to top 1500 frequent words
X = cv.fit_transform(corpus).toarray()   # Features
y = data.iloc[:, 1].values               # Labels (0 = negative, 1 = positive)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=0
)
```

```python
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
```

```python
print("\n---------SCORES--------")
print(f"Accuracy: {accuracy*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap="YlGnBu",
            xticklabels=['Negative','Positive'],
            yticklabels=['Negative','Positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
best_accuracy = 0.0
alpha_val = 0.0
for i in np.arange(0.1, 1.1, 0.1):
    temp_classifier = MultinomialNB(alpha=i)
    temp_classifier.fit(X_train, y_train)
    temp_pred = temp_classifier.predict(X_test)
    score = accuracy_score(y_test, temp_pred)
    print(f"Accuracy for alpha={i:.1f}: {score*100:.2f}%")
    if score > best_accuracy:
        best_accuracy = score
        alpha_val = i

print("\nBest Accuracy: {:.2f}% at alpha={}".format(best_accuracy*100, alpha_val))
```

```python
def predict_sentiment(sample_review):
    # Clean the new review
    sample_review = re.sub('[^a-zA-Z]', ' ', sample_review)
    sample_review = sample_review.lower()
    sample_review_words = sample_review.split()
    sample_review_words = [word for word in sample_review_words if word not in set(stopwords.words('english'))]
    final_review = [ps.stem(word) for word in sample_review_words]
    final_review = ' '.join(final_review)

    # Transform and predict
    temp = cv.transform([final_review]).toarray()
    return classifier.predict(temp)[0]  # returns 0 or 1
```

```python
samples = [
    "The food is really bad.",
    "Food was pretty bad and the service was very slow",
    "The food was absolutely wonderful, from preparation to presentation, very pleasing.",
    "Food average"
]
```

```python
for s in samples:
    sentiment = "Positive" if predict_sentiment(s) else "Negative"
    print(f"Review: {s} -> Sentiment: {sentiment}")
```
